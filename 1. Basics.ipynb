<b>Note: Loc includes both the start and end index, whereas iloc doesn't includes the end index</b>

Choosing between loc and iloc
When choosing or transitioning between loc and iloc, there is one "gotcha" worth keeping in mind, which is that the two methods use slightly different indexing schemes.

iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

Why the change? Remember that loc can index any stdlib type: strings, for example. If we have a DataFrame with index values Apples, ..., Potatoes, ..., and we want to select "all the alphabetical fruit choices between Apples and Potatoes", then it's a lot more convenient to index df.loc['Apples':'Potatoes'] than it is to index something like df.loc['Apples', 'Potatoet] (t coming after s in the alphabet).

This is particularly confusing when the DataFrame index is a simple numerical list, e.g. 0,...,1000. In this case df.iloc[0:1000] will return 1000 entries, while df.loc[0:1000] return 1001 of them! To get 1000 elements using loc, you will need to go one lower and ask for df.loc[0:999].

Otherwise, the semantics of using loc are the same as those for iloc.

## Manipulating index

Index is like an address, that’s how any data point across the dataframe or series can be accessed. Rows and columns both have indexes, rows indices are called as index and for columns its general column names.

Pandas have three data structures dataframe, series & panel. We mostly use dataframe and series and they both use indexes, which make them very convenient to analyse.

```python
data.info()
```

```python
data.set_index('title')
```

```python
data = data.set_index('Unnamed: 0')
print(data)
```

## Conditional Selection

### Using conditions

```python
data.country == 'Italy'
```

We can then use this as a condition to select rows for which we get `True` value.

```python
data.loc[data.country == 'Italy']
```

We can also use multiple conditions

```python
data.loc[(data.country == 'Italy') & (data.points >= 90)]
```

### Using conditional operators

We can use operators built in pandas like `isin`. This helps us select rows which have a column's value in a list of other values.

#### isin

```python
data.loc[data.country.isin(['Italy', 'France'])]
```

#### isnull and notnull

Another important function is the `isnull` and it companion `notnull`. Helps us to check for NaN values in a dataframe

```python
data.loc[data.price.isnull()]
```
