# Renaming

`rename()` lets you rename index or column values by specifying a index or column keyword parameter, respectively. It supports a variety of input formats, but usually a Python dictionary is the most convenient.

````python
import pandas as pd

data_path = 'data/winemag-data-130k-v2.csv'
````

````python
data = pd.read_csv(data_path, index_col='Unnamed: 0')
````

````python
data.head()
````

````python
data.rename(columns={'points':'score'})
````

````python
data.rename(index={0: 'First Entry', 1: 'Second Entry'})
````

You'll probably rename columns very often, but rename index values very rarely. For that, set_index() is usually more convenient.

Both the row index and the column index can have their own name attribute. The complimentary rename_axis() method may be used to change these names

````python
data.rename_axis('wines', axis='rows').rename_axis('fields', axis='columns')
````

# Combining

 Pandas has three core methods for doing this. In order of increasing complexity, these are concat(), join(), and merge(). Most of what merge() can do can also be done more simply with join().

## concat()

Given a list of elements, this function will smush those elements together along an axis.

This is useful when we have data in different DataFrame or Series objects but having the same fields (columns). 

For example: Youtube video dataset from different regions. All the dataset from different regions will have similar columns like `viedoid`, `title`, `channel name`, `publish time`, etc


```(python)
pd.concat([youtube_data_uk, youtube_data_canda])
```

## join()

join() lets you combine different DataFrame objects which have an index in common.

```(python)
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')```


The `lsuffix` and `rsuffix` parameters are necessary here because the data has the same column names in both British and Canadian datasets. If this wasn't true (because, say, we'd renamed them beforehand) we wouldn't need them.

Here we will get a dataframe which has columns from both the dataset but each column from canadian dataset will have the suffix '_CAN' and each column from uk dataset will have the '_UK' suffix. These columns will be respectively filled with NaN values as the dataset don't match
